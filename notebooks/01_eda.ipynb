# Tasks

1. Idenfity Missing Values
2. Fill Title with 'Unknown' and Description with 'No Description Available'
3. Drop rows where rating and release_year is missing
4. Add boolean column 'IsMovie' to identify if it is movie or show
5. Replace the shows having column 'season' value as NA with 1 
6. Save final dataframe as csv "titles_final.csv"

### 1. Idenfiy Missing Values

```python
df.isna().sum()
```

### 2. Fill Title and Description with 'Unknown'

```python
df.fillna({'title': 'Unknown', 'description': 'No Description Available'}, inplace=True)
```

### 3. Drop Rows where imdb_score and release_year is missing

```python
df.dropna(subset=['imdb_score', 'release_year'], inplace=True)
```

### 4. Add boolean column 'IsMovie' to identify if it is movie or show

```python
df['IsMovie'] = np.where(df['type'] == 'MOVIE', True, False)
```

### 5. Replace the shows having column 'season' value as NA with 1

```python
df['seasons'] = np.where((df['type'] == 'SHOW') & (df['seasons'].isna()), 1, df['seasons'])
```

### 6. Save final dataframe as csv "titles_final.csv"

```python
df.to_csv('../data/titles_final.csv', index=False)
```
